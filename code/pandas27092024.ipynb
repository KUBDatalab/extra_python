```python
# store the titles
titles = []

for item in text_files:
    title = str(item[:-4])
    titles.append(title)

# store the text content
texts = []
for item in text_files:
    with open ( text_folder / item , 'r', encoding='utf-8-sig') as f:
        text = f.read()
        texts.append(text)

df1 = pd.DataFrame({ 'title': titles, 'text': texts, 'author': 'grimms' })

df1
```

```python
text_folder = Path.cwd() / '../data/txt_files/hca'
text_files = os.listdir(text_folder)


# store the titles
titles = []

for item in text_files:
    title = str(item[:-4])
    titles.append(title)

# store the text content
texts = []
for item in text_files:
    with open ( text_folder / item , 'r', encoding='utf-8-sig') as f:
        text = f.read()
        texts.append(text)

df2 = pd.DataFrame({ 'title': titles, 'text': texts, 'author': 'hca' })

df2
```

```python
df = pd.concat([df1,df2]).reset_index(drop=True)
```

```python
df
```

```python
import re

def clean_text(text):

    # transform to lower case
    text = text.lower()

    text = text.replace('-', ' ')

    word = re.findall(r'\b\S+\b', text)

    clean_text_string = ' '.join(word)

    return clean_text_string

df['clean_text'] = df['text'].apply(clean_text)

```

```python
import re
def calculate_ttr(text):
    
    # Build word lists
    words = text.split()

    # Calculate the total amount of words
    total_tokens = len(words)

    # Calculate the amount of unique types (unique words)
    unique_types = len(set(words))

    # Calculate TTR (Type-Token Ratio) - the ratio between the total amount of words and unique words  
    ttr = unique_types / total_tokens

    return ttr
```

```python
df['ttr'] = df['clean_text'].apply(calculate_ttr)
```

```python
df
```

```python
# Group by 'author' and calculate the mean TTR
mean_ttr = df.groupby('author')['ttr'].mean().reset_index()
mean_ttr
```

```python
df = pd.merge(df, mean_ttr, on='author', suffixes=('_tale', '_author_mean'))
# inspect
df
```

```python
df['ttr_diff'] = df['ttr_tale'] - df['ttr_author_mean']
```

```python
# plot ttr difference vs title.
df.plot(x='title',y='ttr_diff', kind="bar",figsize=(12,8) , title="TTR difference from average")
```
