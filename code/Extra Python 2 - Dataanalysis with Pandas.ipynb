```python
# import the libraries
import os
from pathlib import Path
import pandas as pd
```

```python
# path to text folder 
text_folder = Path.cwd() / '../data/txt_files/grimm'
text_files = os.listdir(text_folder)

# store title names
titles = []
for i in text_files:
    title = str(i[:-4])
    titles.append(title)

# store texts
texts = []
for item in text_files:
    with open( text_folder / item, 'r', encoding='utf-8-sig') as f:
        text = f.read()
        texts.append(text)

df1 = pd.DataFrame({'title': titles,
                  'text':texts,
                  'author': 'grimms'})

df1
```

## Store data in a dataframe

```python
# path to text folder 
text_folder = Path.cwd() / '../data/txt_files/hca'
text_files = os.listdir(text_folder)

titles = []
for i in text_files:
    title = str(i[:-4])
    titles.append(title)

texts = []
for item in text_files:
    with open( text_folder / item, 'r', encoding='utf-8-sig') as f:
        text = f.read()
        texts.append(text)

df2 = pd.DataFrame({'title': titles,
                  'text':texts,
                  'author': 'hca'})

df2
```

## pd.concat()

Appends rows from one dataframe to another dataframe.

```python
df = pd.concat([df1, df2]).reset_index(drop=True)
```

# Application of functions

```python
import re
def clean_text(text):
    
    # Tranform the text to lower case 
    text = text.lower()
    
    text = text.replace('-', ' ')
    
    # Build word lists
    words = re.findall(r'\b\S+\b', text)

    # join word list to string
    clean_text_string = ' '.join(words)

    return clean_text_string
```

```python
df['clean_text'] = df['text'].apply(clean_text)
```

```python
import re
def calculate_ttr(text):
    
    # Build word lists
    words = text.split()

    # Calculate the total amount of words
    total_tokens = len(words)

    # Calculate the amount of unique types (unique words)
    unique_types = len(set(words))

    # Calculate TTR (Type-Token Ratio) - the ratio between the total amount of words and unique words  
    ttr = unique_types / total_tokens

    return ttr
```

```python
df['ttr'] = df['text'].apply(calculate_ttr)
```

# Groupby and summerize data

```python
# Group by 'author' and calculate the mean TTR
mean_ttr = df.groupby('author')['ttr'].mean().reset_index()
mean_ttr
```

```python
# Group by 'author' and count the number of stories
story_count = df.groupby('author')['title'].count().reset_index()
story_count.columns = ['author', 'story_count']
story_count
```

# Combine datasets

```python
df = pd.merge(df, mean_ttr, on='author', suffixes=('_tale', '_author_mean'))
# inspect
df.head(5)
```

```python
# This is a function called 'split_text' 
# It takes a text as input and splits it into paragraphs using periods as the delimiter.



def split_text(row):
    paragraphs = row.split('.')  #  Split the text into paragraphs using periods as separators.
    return paragraphs  # Return the list of paragraphs.


# This line applies the split_text function to each row in the 'text' column of the DataFramen 'df'.
# It creates a new column called 'paragraphs' to store the segmented text.


df['paragraphs'] = df['text'].apply(lambda row: split_text(row))
```
